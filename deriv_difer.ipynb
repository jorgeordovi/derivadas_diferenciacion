```python
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')

f = x**2 + 3*x + 2

print(f)

sp.pretty_print(f)
```

```python
h = sp.symbols('h')
f_limit=sp.limit((f.subs(x,x+h)-f)/h,h,0)

```

```python
sp.pretty_print(f_limit)
```

```python
f_prime_diff=sp.diff(f,x)
sp.pretty_print(f_prime_diff)
```

```python
import matplotlib.pyplot as plt
```

```python
f_lambdified= sp.lambdify(x,f,'numpy')
f_prime_limit_lambdified= sp.lambdify(x,f_limit,'numpy')
f_prime_diff_lambdified= sp.lambdify(x,f_prime_diff,'numpy')

x_vals=np.linspace(-10,10,400)
```

```python
y_vals=f_lambdified(x_vals)
y_prime_limit_vals=f_prime_limit_lambdified(x_vals)
y_prime_diff_vals=f_prime_diff_lambdified(x_vals)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_vals,y_vals,label=r'$f(x)=x**2 + 3*x + 25', color='blue')
plt.plot(x_vals,y_prime_limit_vals,label=r'Derivadas por definicon', color='red',linestyle='--')
plt.plot(x_vals,y_prime_diff_vals,label=r'Derivadas con diff', color='green',linestyle='-.')
plt.title('Funcion derivada')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0,color='black',linewidth=0.5)
plt.axvline(0,color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
